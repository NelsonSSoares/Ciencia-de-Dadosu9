# Gráficos: `Pandas.plot()` e `Matplotlib`

* Seaborn: https://seaborn.pydata.org/
* Matplotlib: https://matplotlib.org/



**Dataset**: https://edsonmelo.com.br/dados/mtcars.csv

* `mpg`: Milhas por Galão (consumo)
* `cyl`: Número de cilíndros
* `disp`: Cilindada (em polegada cúbica)
* `hp`: Cavalos de Potência (HP)
* `drat`: Relação do eixo traseiro
* `wt`: Peso em (1,000 libras)
* `qsec`: Tempo que atinge 400m (1/4 de milha)
* `vs`: Motor (0 = Forma em V, 1 = Reto)
* `am`: Transmissão (0 = Automático, 1 = Manual)
* `gear`: Número de marchas
* `carb`: Número de carburadores

```python
import pandas as pd # para dados
import numpy as np # para arrays

# Biblioteca Gráfica
import matplotlib.pyplot as plt
```

```python
# carregar o dataset "mtcars"
df = pd.read_csv('https://edsonmelo.com.br/dados/mtcars.csv', index_col = 0)
```

```python
# verificar o DataFrame
print('Shape...: ', df.shape)
print('Linhas..: ', df.shape[0])
print('Colunas.: ', df.shape[1])
```

```python
# head()
df.head()
```

```python
# análise exploratória dos dados
df.describe()
```

```python
# mostrar um gráfico de distribuição (histograma)
# para determinar o tipo do gráfico, usamos o parâmetro "kind"
df.hp.plot(kind='hist')
plt.show()
```

```python
df.hp.plot(kind='bar')
plt.show()
```

```python
df.hp.plot(kind='barh')
plt.show()
```

```python
df.hp.plotot(kind='line')
plt.show()
```

```python
df.hp.plot(kind='kde')
plt.show()
```

```python
# customizando um gráfico

# redimensionando o tamanho da figura
plt.figure(figsize=(8,6))

df.hp.plot(kind='kde')
plt.show()
```

```python
plt.figure(figsize=(8,6))

# título
plt.title('Variável HP por densidade', fontdict={'fontsize': 20})

df.hp.plot(kind='kde')
plt.show()
```

```python
# exemplo de gráfico com os dados ordenados
plt.figure(figsize=(8,6))
df.hp.sort_values(ascending=False).plot(kind='barh')
plt.show()
```

```python
plt.figure(figsize=(8,6))

# dados
df.hp.plot(kind='kde')

# título
plt.title('Variável HP por densidade', fontdict={'fontsize': 20})

# rótulos
plt.xlabel('Cavalos de Potência (HP)')
plt.ylabel('Modelos dos Carros')
plt.show()
```

```python
plt.figure(figsize=(8,6))

# dados
df.hp.plot(kind='kde')

# definir fontes
font_t = {'family': 'arial', 'color': 'blue', 'size': 20}
font_x = {'family': 'serif', 'color': 'red', 'size': 15}
font_y = {'family': 'DejaVu Sans', 'color': 'green', 'size': 18}

# título
plt.title('Variável HP por densidade', fontdict=font_t)

# rótulos
plt.xlabel('Cavalos de Potência (HP)', fontdict=font_x)
plt.ylabel('Modelos dos Carros', fontdict=font_y)
plt.grid()
plt.show()
```

```python
# passando de fase.... agora vamos usar a Matplotlib para gerar os gráficos
# gráfico de dispersão (scatter)

font_t = {'family': 'serif', 'color': 'blue', 'size': 20}
font_x = {'family': 'serif', 'color': 'k', 'size': 15}
font_y = {'family': 'serif', 'color': 'k', 'size': 15}

plt.figure(figsize=(8,6))
plt.title('Cavalos X Cilindradas', fontdict=font_t)
plt.xlabel('Cavalos de Potência', fontdict=font_x)
plt.ylabel('cilindrada (em polegada cúbica)', fontdict=font_y)

# gerar o gráfico
# plt.scatter(eixo_x, eixo_y)
plt.scatter(df.hp, df.disp)
plt.grid()

# salvando a figura
plt.savefig(
    'cavalos_cilindrada.png', 
    dpi=300,
    facecolor='w', 
    edgecolor='r',
    orientation='portrait', 
    papertype=None, 
    format=None,
    transparent=True, 
    bbox_inches=None, 
    pad_inches=0.40,
    frameon=None, 
    metadata=None
    )
plt.show()
```

```python
# Seaborn - biblioteca gráfica mais rebuscada
import seaborn as sns
```

```python
# correlação entre 
# x: Milhas por galão (consumo)
# y: Cavalos de Potência
sns.pairplot(df.loc[:, 'mpg':'hp'])
plt.show()
```

```python
sns.pairplot(df.loc[:, 'mpg':'cyl'])
plt.show()
```

```python
# trabalhando com eixos (x, y)
x = np.linspace(0, 30, 500)

plt.figure(figsize=(8,6))
plt.plot(x, x**2, 
          color='green', 
          marker='o', 
          linestyle='dashed',
          linewidth=0.1, 
         #markersize=12
         )
plt.show()
```

```python
# scatter
plt.scatter(df.hp, df.mpg)
plt.show()
```

```python
# agrupamentos
grupo = df.groupby('cyl').mean().reset_index()
grupo
```

```python
plt.bar(grupo['wt'], grupo['hp'])
plt.grid(True)
plt.show()
```

```python
# pizza

y = np.array([35,25,25,15])
plt.figure(figsize=(8,6))

rotulos = ['Carros', 'Motos', 'Bicicletas', 'Ônibus'] # labels
afastamento = [0.1, 0.0, 0.0, 0.2] # distância entre as fatias
cores = ['black', 'y', 'b', '#4CAF50']

plt.pie(
    y, # valores
    labels=rotulos,
    #colors=cores,
    explode=afastamento
    )
plt.legend(y)
plt.show()
```

```python
# gráfico de pilhas (stackplot)
plt.figure(figsize=(8,6))
x = np.linspace(0, 1.5, 100)
labels = [r'$x$', r'$x^2$', r'$x^3$', r'$x^4$']
plt.stackplot(x, x, x**2, x**3, x**4, labels=labels) # funções exponenciais (x^1 até x^4)
plt.legend(loc='upper left') # canto superior esquerdo
plt.show()
```
